```python
def test(a, b):
    r = int(a - (b)*int(a/b))
    aa = a
    bb = b
    while 1:
        r = aa%bb
        if not r:
            break
        aa = b
        bb = r
    print('Найбільший спільний дільник ', a,' і ', b,' – це ',bb)
    
test(86, 22)
test(54, 12)
test(81, 33)
```
